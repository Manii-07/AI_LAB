**Task 1**
**Appending the list**

```python
#append
fruit_list = ["apple", "banana", "watermelon","mango","kiwi"]
fruit_list.append("orange")
print(fruit_list)
```

**Task 2**
**clearing the list**

```python
#clear
fruit_list = ["apple", "banana", "watermelon","mango","kiwi","orange"]
print(fruit_list)
fruit_list.clear()
print(fruit_list)
```

**Task 3**
**Copying the list**

```python
#copy
fruit_list = ["apple", "banana", "watermelon","mango","kiwi","orange"]
print(fruit_list)
new_fruit_list = fruit_list.copy()
print(new_fruit_list)
```

**Task 4**
**Counting the list**

```python
#count
fruit_list = ["apple", "banana", "watermelon","apple","kiwi","mango","kiwi","orange","apple"]
print(fruit_list)
print(fruit_list.count("apple"))
print(fruit_list.count("kiwi"))
```

**Task 5**
**Extending the list**

```python
#extend
fruit_list = ["apple", "banana", "watermelon","mango","kiwi"]
veg_list = ["carrot","potato","onion"]
print(fruit_list)
print(veg_list)
list3 = fruit_list + veg_list
print(list3)
```

**Task 10**

```python
#reversing
fruit_list = ["apple", "banana", "watermelon","mango","kiwi"]
print(fruit_list)
fruit_list.reverse()
print(fruit_list)
```

**Task 11**
**Sorting the list**


```python
#sort
fruit_list = ["apple", "banana", "watermelon","mango","kiwi","orange"]
print(fruit_list)
fruit_list.sort()
print(fruit_list)

car_list = ["Ford", "BMW", "Volvo","Audi"]
print(car_list)
car_list.sort()
print(car_list)

numw_list = [100, 50, 65, 82, 23]
print(numw_list)
numw_list.sort()
print(numw_list)
```

**Task 12**
**counting the Repeted words**


```python
fruit_list = ["apple", "banana", "watermelon", "apple", "kiwi", "mango", "kiwi", "orange", "apple"]
print
repeated = []

for fruit in fruit_list:
    if fruit_list.count(fruit) > 1 and fruit not in repeated:
        repeated.append(fruit)

print("Original list:", fruit_list)
print("Repeated items:", repeated)

```
