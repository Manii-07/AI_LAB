***Task 2***

```python
#Create a 2-D array and a 3-D. Split the 2-D array into three 2-D arrays along rows.
import numpy as np
# 1. Create a 2-D array
array_2d = np.arange(24).reshape(6, 4)
print("2-D Array")
print(array_2d)
print("\n")

# 2. Create a 3-D array
array_3d = np.arange(24).reshape(2, 3, 4)
print("3-D Array")
print(array_3d)
print("\n")


# 3. Split the 2-D array into three 2-D arrays along rows
split_arrays = np.split(array_2d, 3)
print("Splitted Arrays:")
print("Array 1 \n",split_arrays[0])
print("Array 1 \n",split_arrays[1])
print("Array 1 \n",split_arrays[2])
```

```python
#Create a 3x3 matrix with values ranging from 2 to 10, convert array to a float type. Write a python program to find the maximum and minimum value to flattened array.
# 3x3 matrix with values 2â€“10
matrix = np.arange(2, 11).reshape(3, 3)
print("Matrix:\n", matrix)

# Convert to float
matrix_float = matrix.astype(float)
print("Float Matrix:\n", matrix_float)

# Flatten and find max & min
flat = matrix_float.flatten()
print("Flattened array:", flat)
print("Max value:", np.max(flat))
print("Min value:", np.min(flat))
```

```python
import random

# Create a list of 150 random numbers between 1 and 500
random_numbers = [random.randint(1, 100) for _ in range(150)]

# 1. Show all numbers in the list
print("1. Full list of 150 random numbers:")
print(random_numbers)

# 2. Find and show all even numbers
even_numbers = [num for num in random_numbers if num % 2 == 0]
print("\n2. Even numbers in the list:")
print(even_numbers)

# 3. Find and show all prime numbers
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [num for num in random_numbers if is_prime(num)]
print("\n3. Prime numbers in the list:")
print(prime_numbers)

# 4. Function to print multiples of a given number
def find_multiples(numbers_list, given_number):
    print(f"\n4. Multiples of {given_number}:")
    multiples = [num for num in numbers_list if num % given_number == 0]

    if multiples:
        print(multiples)
    else:
        print(f"No multiples of {given_number} were found in the list.")

find_multiples(random_numbers, 5)
```
