***Task 01***


```python
import numpy as np

arr = np.arange(1, 16)
split_arrays = np.array_split(arr, 3)
print("Original Array:", arr)
print("Split Arrays:", split_arrays)

total_sum = sum(a.sum() for a in split_arrays)
print("Sum of all elements:", total_sum)

```

```python
filtered = arr[(arr > 5) & (arr < 15)]
print("Numbers >5 and <15:", filtered)
```

***Task 03***


```python
set1 = {0, 2, 4, 6, 8, 10}
set2 = set(range(1, 11))

print("Set1 (Even numbers):", set1)
print("Set2 (Positive integers):", set2)

set2_list = list(set2)
set2_list.sort()
set2_list.pop(1)
set2_list.pop(3)
set2 = set(set2_list)
print("Set2 after removing 2nd and 5th element:", set2)

set3 = set1 & set2
print("Intersection of Set1 and Set2:", set3)
```

***Task 05***


```python
cricket = {"Ali", "haider", "Usman", "Hadia"}
football = {"Haider", "Hadia", "Zara", "Ahmed"}

print("Cricket:", cricket)
print("Football:", football)
print("Common players:", cricket & football)
print("Only cricket players:", cricket - football)
print("Only football players:", football - cricket)
```

***Task 06***


```python
students = {
    "SAP101": {"sapid": "SAP101", "name": "Ali", "age": 19, "semester": "Six"},
    "SAP102": {"sapid": "SAP102", "name": "Haider", "age": 21, "semester": "Six"},
    "SAP103": {"sapid": "SAP103", "name": "Usman", "age": 21, "semester": "Six"},
    "SAP104": {"sapid": "SAP104", "name": "Hadia", "age": 22, "semester": "Six"},
    "SAP105": {"sapid": "SAP105", "name": "alisha", "age": 23, "semester": "Six"},
}

print("\nOriginal Students Dictionary:")
for k, v in students.items():
    print(k, ":", v)

# Update semester to Seven
for record in students.values():
    record["semester"] = "Seven"

print("\nUpdated Students Dictionary:")
for k, v in students.items():
    print(k, ":", v)
```
